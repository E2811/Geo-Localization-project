```python
import pandas as pd
import functions as F
dist = pd.read_csv('../OUTPUT/distances.csv')
```

```python
print('Rate your preferences when choosing your office location')
rate_v = int(input("Rate from 1 to 5 having vegan restaurants near the office:"))
rate_s = int(input("Rate from 1 to 5 having schools near the office:"))
rate_a = int(input("Rate from 1 to 5 having an airport near the office:"))
rate_sb = int(input("Rate from 1 to 5 having starbucks near the office:"))
```

```python
rating = [rate_s, rate_v, rate_a, rate_sb]
for e,r in enumerate(rating):
    dist[f'distance{e+1}']= dist[f'distance{e+1}'].apply(lambda x: F.ponderation(r,x))
```

```python
dist['Total'] =  dist.iloc[:, -4:-1].sum(axis=1)
```

```python
dist.sort_values('Total',ascending=True)
```
